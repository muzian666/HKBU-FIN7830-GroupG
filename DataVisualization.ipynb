# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the visualization style
sns.set(style="whitegrid")

# Create a list of numeric and categorical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Create subplots for numeric data
fig, axes = plt.subplots(len(numeric_cols)//3, 3, figsize=(18, 24))

for col, ax in zip(numeric_cols, axes.flatten()):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Create bar plots for categorical data
fig, axes = plt.subplots(len(categorical_cols), 1, figsize=(18, 24))

for col, ax in zip(categorical_cols, axes.flatten()):
    sns.countplot(data=df, x=col, ax=ax)
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Explore the relationship between numerical features and the target variable 'Attrition'
fig, axes = plt.subplots(len(numeric_cols)//3, 3, figsize=(18, 24))

for col, ax in zip(numeric_cols, axes.flatten()):
    sns.boxplot(data=df, x='Attrition', y=col, ax=ax)
    ax.set_title(f'{col} vs Attrition')
    ax.set_xlabel('Attrition')
    ax.set_ylabel(col)

plt.tight_layout()
plt.show()

# Explore the relationship between categorical features and the target variable 'Attrition'
fig, axes = plt.subplots(len(categorical_cols)-1, 1, figsize=(18, 24))  # Exclude 'Attrition' itself

for col, ax in zip(categorical_cols[:-1], axes.flatten()):  # Exclude 'Attrition' itself
    sns.countplot(data=df, x=col, hue='Attrition', ax=ax)
    ax.set_title(f'{col} vs Attrition')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```
